```python
response = requests.get(
url='https://proxy.scrapeops.io/v1/',
params={
'api_key': 'e2d639dd-d209-4588-b225-05ea12b07cb9',
'url': 'https://www.indeed.com/cmp/US-Department-of-the-Treasury/reviews',
}
)
print('body: ', response.content)
```

```python
soup = BS(response.text)
```

```python
soup
```

```python
department_name_element = soup.find('div', {'class': 'css-19rjr9w', 'itemprop': 'name'})
department_name = department_name_element.get_text(strip=True)
```

```python
overall_rating_element = soup.find('span', {'class': 'css-htn3vt', 'itemprop': 'ratingValue'})
overall_rating_value = overall_rating_element.get_text(strip=True)
```

```python
sub_rating_elements = soup.select('.css-1vmx0e0')
sub_rating_values = [element.get_text() for element in sub_rating_elements]
```

```python
department_name
```

```python
overall_rating_value
```

```python
sub_rating_values
```

```python
category_ratings = {
    'Work-Life Balance': {'Rating': sub_rating_values[0]},
    'Pay & Benefits': {'Rating': sub_rating_values[1]},
    'Job Security & Advancement': {'Rating': sub_rating_values[2]},
    'Management': {'Rating': sub_rating_values[3]},
    'Culture': {'Rating': sub_rating_values[4]},
}

# Add department_name to each category
for category, details in category_ratings.items():
    details['Department'] = department_name
    
category_ratings['Overall Rating'] = {'Department': department_name, 'Rating': overall_rating_value}

# Reshape the data
data = [{'Department': details['Department'], 'Category': category, 'Rating': details['Rating']} for category, details in category_ratings.items()]

# Convert to DataFrame
df = pd.DataFrame(data)

df
```
