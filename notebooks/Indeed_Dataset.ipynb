```python
path = r'C:\Users\AaronH\Documents\nss_data_analytics\Projects\federal_employee_job_satisfaction\data'
filenames = [file for file in os.listdir(path) if file.endswith('.xlsx')]

df = pd.concat([pd.read_excel(os.path.join(path, file)) for file in filenames], ignore_index=True)
```

```python
df_final = df.drop(['Unnamed: 0'], axis=1)
```

```python
df_final
```

```python
df_melted = pd.melt(df_final, id_vars=['Department', 'Category'], value_vars='Rating')
```

```python
df_reshape = df_melted.pivot(index='Department', columns='Category', values='value').reset_index()
```
